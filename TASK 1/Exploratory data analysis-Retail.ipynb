# EXPLORATORY DATA ANALYSIS

```python
import seaborn as sns
```

```python
sns.pairplot(dt)
```

<b> Heatmap visulization 

```python
correlation=data.corr()
plt.figure(figsize=(15,15))
sns.set(font_scale=2)
sns.heatmap(correlation,annot=True,cmap="YlGnBu")
plt.show()
```

```python
#obsrving the categories
plt.figure(figsize=(8,8))
plt.title('category')
plt.pie(data["Category"].value_counts(),labels=data['Category'].value_counts().index,autopct="%1.1f%%")
plt.show()
```

```python
data.groupby('Category')["Profit","Sales"]
```

```python
plt.figure(figsize=(30,15))
data.groupby('Category')["Profit","Sales"].agg(['sum']).plot.bar()
plt.ylabel("Profit")
plt.show()
```

```python
#computing top categories in term of sales from first 200 observations
topSale=data.groupby("Category").Sales.sum().nlargest(n=200)
#computing top categories in term of Profit from first 200 observations
topProfit=data.groupby("Category").Profit.sum().nlargest(n=200)
#plotting to see it visually
plt.style.use('seaborn')
topSale.plot(kind="bar",figsize=(15,10),fontsize=18,color="blue")
topProfit.plot(kind="bar",figsize=(15,10),fontsize=18,color="red")
plt.xlabel("Category",fontsize=17)
plt.ylabel("TotalSale/Profit",fontsize=17)
plt.title(" Category Sale vs Profit",fontsize=17)
plt.show()
```

```python
#computing top subcategories in term of sales from first 200 observations
topsubSale=data.groupby("Sub-Category").Sales.sum().nlargest(n=200)
#computing top subcategories in term of Profit from first 200 observations
topsubProfit=data.groupby("Sub-Category").Profit.sum().nlargest(n=200)
#plotting to see it visually
plt.style.use('seaborn')
topsubSale.plot(kind="bar",figsize=(15,10),fontsize=18,color="blue")
topsubProfit.plot(kind="bar",figsize=(15,10),fontsize=18,color="green")
plt.xlabel("Sub-Category",fontsize=17)
plt.ylabel("TotalSale/Profit",fontsize=17)
plt.title("SubCategory Sale vs Profit",fontsize=17)
plt.show()
```

<b>  This Shows Copiers and phones have high sale

```python
plt.figure(figsize=(30,15))
substate=data.groupby(["Sub-Category"])['Profit'].sum().nlargest(100)
substate.plot.bar(color="green")
```

# Observing the discount

```python
sns.lineplot(data["Discount"],data["Profit"],data=data,color="blue")
```

# Visualization of  sale in Regions

```python
plt.title("Region")
plt.pie(data["Region"].value_counts(),labels=data["Region"].value_counts().index,autopct="%1.1f%%")
plt.show()
```

<b>The above graph clearly indicates that:
west>East>Central>south :Order of sales in different region.<b>

```python
#computing top state in term of sales from first 200 observations
StateS=data.groupby("State").Sales.sum().nlargest(n=10)
#computing top state in term of Profit from first 200 observations
StateP=data.groupby("State").Profit.sum().nlargest(n=10)
#plotting to see it visually
plt.style.use('seaborn')
StateS.plot(kind="bar",figsize=(15,10),fontsize=18,color="blue")
StateP.plot(kind="bar",figsize=(15,10),fontsize=18,color="yellow")
plt.xlabel("States",fontsize=17)
plt.ylabel("TotalSale",fontsize=17)
plt.title("Top10 State Sale vs Profit",fontsize=17)
plt.show()
```

```python
plt.style.use("seaborn")
data.plot(kind="scatter",figsize=(15,15),x="Sales",y="Profit",c="Discount",colormap="jet")
plt.ylabel("Totalprofits",fontsize=15)
plt.title("Relation between sale profit and discount")
plt.show()
```
